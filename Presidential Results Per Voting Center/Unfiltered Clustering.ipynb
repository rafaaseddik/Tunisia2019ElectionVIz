```python
sortedNames = sorted(names)
separated = [i.replace(".xlsx","").split("-") for i in names]
```

```python
cache= separated
result=[]
while len(cache)>0:
    firstElt = cache[0]
    result.append([firstElt])
    print("------------"," ".join(firstElt))
    found = []
    for j in range(1,len(cache)):
        secondElt = cache[j]
        if secondElt[-1]==firstElt[0] or secondElt[0]==firstElt[0] or secondElt[-1]==firstElt[-1]or secondElt[0]==firstElt[-1] :
            print(" ".join(secondElt))
            result[-1].append(secondElt)
            found.append(j)
    found = sorted(found,reverse=True)
    for j in found:
        del cache[j]
    cache.pop(0)
        
        
```
