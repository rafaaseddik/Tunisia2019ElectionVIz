```python
from selenium import webdriver
browser = webdriver.Chrome()
```

```python
browser.get("https://majles.marsad.tn/2014/fr/assemblee")
```

```python
cards = browser.find_elements_by_css_selector(".for-elus .card")
deputies = []
```

```python
class Deputy:
    def __init__(self,name,url,age,siege,profession,circ,photo_url,electoral_list,tourism):
        self.name = name
        self.url = url
        self.age = age
        self.siege = siege
        self.profession = profession
        self.circ = circ
        self.photo_url = photo_url
        self.electoral_list = electoral_list
        self.tourism = tourism
```

```python
class Movement:
    def __init__(self,parti,startDate,endDate):
        self.parti=parti
        self.startDate=parseDate(startDate)
        self.endDate =parseDate(endDate)
```

```python
def parseDate(fr):
    if(fr=="Maintenant"):
        return "2019-10-06"
    mounth,year=fr.split(" ")
    dictio = {
    "janvier":"01",
    "février":"02",
    "mars":"03",
    "avril":"04",
    "mai":"05",
    "juin":"06",
    "juillet":"07",
    "août":"08",
    "septembre":"09",
    "octobre":"10",
    "novembre":"11",
    "décembre":"12"
    }
    return year+"-"+dictio[mounth]+"-01";
```

```python
deputies = {}
for card in cards:
    name = card.get_attribute("data-nom")
    url = card.find_element_by_class_name("link-elu").get_attribute("href")
    age = card.get_attribute("data-age")
    siege = card.get_attribute("data-siege")
    profession = card.get_attribute("data-profession")
    circ = card.get_attribute("data-circ")
    photo_url = card.get_attribute("data-photo").replace(".thumb50","")
    electoral_list = card.get_attribute("data-liste")
    deputies[str(siege)] = {
        "name":name,
        "url":url,
        "age":age,
        "siege":siege,
        "profession":profession,
        "circ":circ,
        "photo_url":photo_url,
        "electoral_list":electoral_list
    }
```

```python
i=0
for siege in deputies:
    browser.get(deputies[siege]["url"]);
    if("tourism" not in deputies[siege]):
        deputies[siege]["tourism"] = get_movement()
    i=i+1
    print("----------",i,"/216")
```

```python
def get_movement():
    result= [];
    movementBox = browser.find_element_by_xpath('//*[@id="main"]/div[3]/div[2]/div/div[1]')
    movements = movementBox.find_elements_by_css_selector(".list-group li")
    for m in movements:
        movement_title = m.find_element_by_css_selector('span').text
        if(movement_title == "Aucun bloc"):
            return [{
                "title":movement_title,
                "startDate":"2014-12-01",
                "endDate":parseDate("Maintenant")
            }]
            
        movement_period = m.find_element_by_css_selector('.text-muted').text
        movement_period_start,movement_period_end = movement_period.split("→")
        movement = Movement(movement_title,movement_period_start.strip(),movement_period_end.strip())
        result.append({
            "title":movement_title,
            "startDate":parseDate(movement_period_start.strip()),
            "endDate":parseDate(movement_period_end.strip())
        })
    return result;
    
```

```python
outfile = open('Deputiesq_final.json',"w+",encoding="utf-8")
outfile.write(str(list(deputies.values())).replace('\'','"').replace("/2014/uploads/images","https://majles.marsad.tn/2014/uploads/images"))
```
